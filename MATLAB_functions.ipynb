```matlab
y = sin(x)
```

#### This function gives you the minimum value of y

```matlab
a = min(y)
```

#### Here, the capital I will tell you the location of the minimum value of y (the min function can also give you location)

```matlab
[a, I] = min(y)
```

#### 3D plots

```matlab
[x, y, z] = cylinder (10:-1:0, 50);
surf (x, y, z);
title ("a cone");
```

```matlab
[x, y, z] = sphere (40);
surf (3*x, 3*y, 3*z);
axis equal;
title ("sphere of radius 3");
```

#### Using teapot geometry (built in MATLAB feature)

```matlab
[verts, faces, cindex] = teapotGeometry;
figure
p = patch('Faces',faces,'Vertices',verts,'FaceVertexCData',cindex,'FaceColor','interp')

```

```matlab
figure
p = patch('Faces',faces,'Vertices',verts,'FaceVertexCData',cindex,'FaceColor','interp')
view(-151,30)     % change the orientation
```
